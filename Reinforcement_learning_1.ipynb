```python
import gymnasium
from huggingface_sb3 import load_from_hub, package_to_hub
from huggingface_hub import notebook_login

from stable_baselines3 import PPO
from stable_baselines3.common.env_util import make_vec_env
from stable_baselines3.common.evaluation import evaluate_policy
from stable_baselines3.common.monitor import Monitor
```

```python
#make env
import gymnasium as gym
env=gym.make("LunarLander-v3")  #step2 = env=gym.make-- makes the env ("env name")

observation,info=env.reset() #step 3 initiate state

for i in range(10):
  action=env.action_space.sample()
  print("action taken:",action)
observation,reward,terminated,truncated,info=env.step(action)
if terminated or truncated:
  print("environment is rest")
  observation,info=env.reset()
```

```python
import pandas as pd
```

```python
env=gym.make("LunarLander-v3")
env.reset()
print('__________observation space__________\n')
print("observation space shape",env.observation_space.shape)

observation_sample = env.observation_space.sample()
observation_labels = [
    "Horizontal pad coordinate (x)",
    "Vertical pad coordinate (y)",
    "Horizontal speed (x)",
    "Vertical speed (y)",
    "Angle",
    "Angular speed",
    "If the left leg contact point has touched the land (boolean)",
    "If the right leg contact point has touched the land (boolean)"
]
labeled_observation = pd.Series(observation_sample, index=observation_labels)
print("sample observation\n", labeled_observation)
```

```python
print("____action space____\n")
print("action space shape",env.action_space.n)
print("action space sample",env.action_space.sample())
```

```python
env=make_vec_env('LunarLander-v3',n_envs=16)
```

```python
env=gym.make('LunarLander-v3')
model=PPO(
    policy="MlpPolicy",
    env=env,
    n_steps=1024,
    batch_size=64,
    n_epochs=4,
    gamma=0.999,
    gae_lambda=0.9,
    verbose=1
)
model.learn(total_timesteps=int(2e5))
```

```python
model_name="ppo-LunarLander-v3"
model.save(model_name)
```

```python
eval_env=Monitor(gym.make("LunarLander-v3",render_mode="rgb_array"))
mean_reward,std_reward=evaluate_policy(model,eval_env,n_eval_episodes=10,deterministic=True)
print(f"mean_reward={mean_reward:.2f} +/- {std_reward}")
```

```python
import gymnasium as gym
from stable_baselines3.common.vec_env import DummyVecEnv, VecVideoRecorder
from stable_baselines3.common.env_util import make_vec_env
from stable_baselines3.common.monitor import Monitor
from huggingface_sb3 import package_to_hub
import os

env_id="LunarLander-v3"
model_architecture="PPO"
repo_id="venky0310/ppo-LunarLander-v3"
commit_message="trained agent PPO LunarLander-v3"

# Wrap the evaluation environment with Monitor and VecVideoRecorder
eval_env = DummyVecEnv([lambda: Monitor(gym.make(env_id, render_mode="rgb_array"))])
video_folder = "videos"
video_length = 1000
eval_env = VecVideoRecorder(eval_env, video_folder=video_folder, record_video_trigger=lambda step: step == 0, video_length=video_length, sync_encoder=False)


package_to_hub(model=model,
                model_name=model_name,
                model_architecture=model_architecture,
                env_id=env_id,
                eval_env=eval_env,
                repo_id=repo_id,
                commit_message=commit_message)
```
